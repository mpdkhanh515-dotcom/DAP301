```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('/content/DATASET.xlsx')

# Ensure we have numeric and categorical columns
numeric_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(exclude='number').columns.tolist()

# 1. Histogram
plt.figure(); df[numeric_cols[0]].hist(); plt.title("Histogram"); plt.show()

# 3. Scatter
plt.figure(); plt.scatter(df[numeric_cols[0]], df[numeric_cols[1]]); plt.title("Scatter"); plt.show()

# 4. Line chart
plt.figure(); plt.plot(df[numeric_cols[0]].reset_index(drop=True)); plt.title("Line Chart"); plt.show()


# 7. Heatmap correlation
plt.figure(); plt.imshow(df[numeric_cols].corr()); plt.title("Heatmap"); plt.colorbar(); plt.show()

# 8. KDE
plt.figure(); df[numeric_cols[0]].plot(kind='kde', title="KDE"); plt.show()

# 10. Area chart
plt.figure(); df[numeric_cols[0]].plot(kind='area', title="Area Chart"); plt.show()

# 11. Bubble
plt.figure(); plt.scatter(df[numeric_cols[0]], df[numeric_cols[1]], s=df[numeric_cols[0]]*10); plt.title("Bubble"); plt.show()

# 12. Violin
plt.figure(); plt.violinplot(df[numeric_cols[0]].dropna()); plt.title("Violin"); plt.show()

# 13. Step
plt.figure(); plt.step(range(len(df)), df[numeric_cols[0]]); plt.title("Step Chart"); plt.show()

# 14. ECDF
plt.figure(); sorted_vals = df[numeric_cols[0]].sort_values().reset_index(drop=True)
plt.plot(sorted_vals, (sorted_vals.rank(method="first")/len(sorted_vals))); plt.title("ECDF"); plt.show()



# 17. Swarm-like (jittered scatter)
import numpy as np
plt.figure(); x = np.random.normal(0,0.04,len(df))
plt.scatter(x, df[numeric_cols[0]]); plt.title("Swarm-like"); plt.show()

# 18. Radar chart
labels = numeric_cols[:3]
values = df[labels].mean().tolist()
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values += values[:1]; angles += angles[:1]
fig = plt.figure(); ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values); ax.fill(angles, values, alpha=0.1)
ax.set_xticks(angles[:-1]); ax.set_xticklabels(labels); plt.title("Radar"); plt.show()



```
